```python
import os
import pandas as pd
import re

def parse_chat_group_name(file_name, team_name):
    # Remove the "WhatsApp Chat with " prefix and the file extension
    group_name_with_extension = file_name.replace("WhatsApp Chat with ", "")
    group_name = os.path.splitext(group_name_with_extension)[0]
    
    # Check for duplicates like (1), (2) etc., and remove them
    group_name = re.sub(r'\(\d+\)$', '', group_name)
    print(group_name)

    # Special rule for 'SALES' team
    if team_name.lower() == 'sales':
        # Check for specific patterns after an underscore or the bracketed date format
        pattern = re.compile(r'.+?_(EDOOFA|edoofa|Edoofa|EA)|\(\d{2}_\d{2}\)')
        expected_format = 'yes' if pattern.search(group_name) else 'no'
    else:
        # Check for special characters in the first two parts
        parts = group_name.split()
        print(parts)
        if len(parts) >= 2 and not (re.search(r'[^A-Za-z0-9]', parts[0]) or re.search(r'[^A-Za-z0-9]', parts[1])):
            expected_format = 'yes'
        else:
            expected_format = 'no'

    return group_name, expected_format


def fetch_chat_data(date_directory):
    # Define the columns for the DataFrame
    columns = ['Date', 'Team', 'Person', 'Chat Name', 'Chat Group Name', 'Expected Format', 'File Size']
    data = []

    # Iterate through each 'Date' directory
    for date_folder in os.listdir(date_directory):
        date_path = os.path.join(date_directory, date_folder)
        if os.path.isdir(date_path):
            # Iterate through each 'Team' subdirectory
            for team_folder in os.listdir(date_path):
                team_path = os.path.join(date_path, team_folder)
                if os.path.isdir(team_path):
                    # Iterate through each 'Person' subdirectory
                    for person_folder in os.listdir(team_path):
                        person_path = os.path.join(team_path, person_folder)
                        if os.path.isdir(person_path):
                            # Process each chat file in the 'Person' subdirectory
                            for file in os.listdir(person_path):
                                if file.endswith('.txt'):  # Assuming chat files are .txt files
                                    file_path = os.path.join(person_path, file)
                                    file_size = os.path.getsize(file_path)

                                    # Extract chat group name and check format
                                    chat_group_name, expected_format = parse_chat_group_name(file, team_folder)



                                    # Add the extracted details to the data list
                                    data.append([date_folder, team_folder, person_folder, file, chat_group_name, expected_format, file_size])

    # Create a DataFrame from the data list
    df = pd.DataFrame(data, columns=columns)
    return df

# Replace 'your_date_directory_path' with the path to the 'Date' directory
date_directory = 'C:\\Users\\mauriceyeng\\Python\\Daily-Reports\\Test\\V1_maurice\\TestingData'
chat_data_df = fetch_chat_data(date_directory)

# Display the DataFrame
chat_data_df.head(10)

```

```python
import os
import pandas as pd
import datetime
import re

# Counselor names for each team
counselor_names = {
    'sales': ['sahil', 'shubham', 'austin', 'sagar', 'kunal', 'shashwat', 'harmehak', 'arshita', 'tushti', 'pallika'],
    'kam': ['kirti', 'shivjeet', 'milan', 'ashi'],
    'ewyl': ['saloni', 'aditi', 'sharda', 'jasmine', 'ananya']
}

# Function to map a message's timestamp to a 5-minute interval
def get_time_interval(timestamp):
    time_obj = datetime.datetime.strptime(timestamp, '%d/%m/%y, %I:%M %p')
    interval = time_obj.replace(second=0, microsecond=0, minute=5 * (time_obj.minute // 5))
    return interval

# Function to process chat files for a person and a date
def process_chat_files(date_directory, team_folder, person_folder):
    columns = [datetime.time(hour=h, minute=m).strftime('%I:%M %p') for h in range(24) for m in range(0, 60, 5)]
    df = pd.DataFrame(0, index=['counselor', 'student'], columns=columns)

    person_found = False
    person_path = os.path.join(date_directory, team_folder, person_folder)

    # Debug: Check if the person's directory exists
    if not os.path.isdir(person_path):
        print(f"Directory not found for: {person_path}")
        return None

    print(f"Processing chat files in: {person_path}")  # Debug

    for file in os.listdir(person_path):
        if file.endswith('.txt'):
            file_path = os.path.join(person_path, file)
            print(f"Reading file: {file_path}")  # Debug
            with open(file_path, 'r', encoding='utf-8') as chat_file:
                for line in chat_file:
                    match = re.match(r'(\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2} [ap]m) - (.*?): (.*)', line)
                    if match:
                        timestamp, sender, message = match.groups()
                        interval = get_time_interval(timestamp).strftime('%I:%M %p')
                        sender_lower = sender.lower()
                        if any(counselor.lower() in sender_lower for counselor in counselor_names[team_folder.lower()]):
                            df.at['counselor', interval] = 1
                            person_found = True
                        else:
                            df.at['student', interval] = 1

    if not person_found:
        print(f"No counselor messages found in files for: {person_folder}")

    return df if person_found else None

# Main Program
root_directory = 'C:\\Users\\mauriceyeng\\Python\\Daily-Reports\\Test\\V1_maurice\\TestingData'  # Replace with the actual path

df_dict = {}  # Dictionary to store DataFrames

# Navigate through the directory structure
for date_folder in os.listdir(root_directory):
    date_path = os.path.join(root_directory, date_folder)
    if os.path.isdir(date_path):
        for team_folder in os.listdir(date_path):
            team_path = os.path.join(date_path, team_folder)
            if os.path.isdir(team_path):
                for person_folder in os.listdir(team_path):
                    df_person_date = process_chat_files(date_path, team_folder, person_folder)
                    if df_person_date is not None:
                        df_key = f"df_{team_folder.lower()}_{person_folder.lower()}"
                        df_dict[df_key] = df_person_date
                        print(f"DataFrame for {person_folder} on {date_folder} created.")
                    else:
                        print(f"No chat data found for {person_folder} on {date_folder}.")

# At this point, df_dict contains all the DataFrames for each person and date
```
