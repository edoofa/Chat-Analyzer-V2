```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data generation
np.random.seed(0)
time_intervals = pd.date_range("00:00", "23:59", freq="5T").strftime('%I:%M %p')
data = {
    "Person": np.random.randint(0, 5, len(time_intervals)),
    "Others": np.random.randint(0, 5, len(time_intervals))
}
df_sample = pd.DataFrame(data, index=time_intervals)

# 1. Line Graph
plt.figure(figsize=(12, 6))
plt.plot(df_sample.index, df_sample['Person'], label='Person', marker='o')
plt.plot(df_sample.index, df_sample['Others'], label='Others', marker='x')
plt.xticks(ticks=df_sample.index[::12], rotation=45)
plt.xlabel('Time')
plt.ylabel('Number of Messages')
plt.title('Message Frequency Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Bar Chart
plt.figure(figsize=(6, 4))
df_sample.sum().plot(kind='bar')
plt.xlabel('Sender')
plt.ylabel('Total Messages')
plt.title('Total Messages by Person vs Others')
plt.tight_layout()
plt.show()

# 3. Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_sample.T, cmap="YlGnBu", linewidths=.5)
plt.xlabel('Time')
plt.ylabel('Sender')
plt.title('Heatmap of Message Frequency')
plt.tight_layout()
plt.show()
```

```python
# Continuing from the previous sample data...

# 4. Pie Chart
plt.figure(figsize=(6, 6))
df_sample.sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Messages by Person vs Others')
plt.ylabel('')  # Hide the y-label
plt.show()

# 5. Stacked Bar Chart
plt.figure(figsize=(12, 6))
df_sample.plot(kind='bar', stacked=True)
plt.xlabel('Time')
plt.ylabel('Number of Messages')
plt.title('Stacked Bar Chart of Messages Over Time')
plt.xticks(ticks=range(0, len(df_sample.index), 12), labels=df_sample.index[::12], rotation=45)
plt.tight_layout()
plt.show()

# 6. Scatter Plot (Example: Comparing 'Person' and 'Others' messages)
plt.figure(figsize=(8, 6))
plt.scatter(df_sample.index, df_sample['Person'], label='Person', alpha=0.5)
plt.scatter(df_sample.index, df_sample['Others'], label='Others', alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Number of Messages')
plt.title('Scatter Plot of Messages Over Time')
plt.legend()
plt.xticks(ticks=df_sample.index[::12], rotation=45)
plt.tight_layout()
plt.show()
```
